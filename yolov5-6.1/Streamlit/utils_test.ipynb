```python
import os
import cv2 as cv

def get_file_list(suffix , path):
    #intput_template_all = []
    intput_template_all_path = []

    for root, dir, file in os.walk(path, topdown=False):
        for name in file:
            if name.split('.')[1] == suffix:
                intput_template_all_path.append(os.path.join(root, name))

    return intput_template_all_path

file_list = get_file_list("png", r"D:\个人信息\获奖情况")

for file in file_list:
    print(file)
    img = cv.imread(file)
    if img is None:
        print(f"无法加载图像: {file}")
        continue

    # 显示图像
    cv.imshow("Image", img)
    cv.waitKey(0)  # 等待用户按键
    cv.destroyAllWindows()  # 关闭所有窗口
```
